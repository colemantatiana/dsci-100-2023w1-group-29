```r
library(tidyverse)
library(repr)
library(readxl)
library(tidymodels)
library(dplyr)
```

```r
set.seed(29)
file_path <- "https://github.com/colemantatiana/dsci-100-2023w1-group-29/raw/main/seeds.csv"
seeds <- read_delim(file_path, col_names = FALSE)
seeds <- select(seeds, X2:X9) |>
    mutate(X9 = as_factor(X9))
colnames(seeds) <- c("area", "perimeter", "compactness", 
                     "kernel_length", "kernel_width", "asymmetry_coefficient", 
                     "kernel_groove", "type")
summary(seeds)
seeds_plot <- seeds |>
  ggplot(aes(x = compactness, 
             y = kernel_groove, 
             colour = type)) +
  geom_point()
# View preliminary plot
seeds_plot
```

```r
# Create training (75%) and testing (25%) data split 
seeds_split <- initial_split(seeds, prop = 0.75, strata = type)  
seeds_train <- training(seeds_split)   
seeds_test <- testing(seeds_split)
# View training and testing stes
# seeds_train
# seeds_test

# Visualize training data
seeds_train_plot <- seeds_train |>
  ggplot(aes(x= compactness, 
             y= kernel_groove, 
             colour = type)) +
  geom_point()
seeds_train_plot

seeds_recipe <- recipe(type ~ compactness + kernel_groove , data = seeds_train) |>
   step_scale(all_predictors()) |>
   step_center(all_predictors())
seeds_recipe

knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = 3) |>
      set_engine("kknn") |>
      set_mode("classification")

seeds_fit <- workflow() |>
      add_recipe(seeds_recipe) |>
      add_model(knn_spec) |>
      fit(data = seeds_train)

seeds_fit

seeds_test_predictions <- predict(seeds_fit, seeds_test) |>
      bind_cols(seeds_test)
seeds_test_predictions

seeds_fit_accuracy <- seeds_test_predictions |>
        metrics(truth = type, estimate = .pred_class)    
seeds_fit_accuracy

seeds_conf <- seeds_test_predictions |> 
      conf_mat(truth = type, estimate = .pred_class)
seeds_conf

seeds_vfold <- vfold_cv(seeds_train, v = 5, strata = type)

seeds_resample_fit <- workflow() |>
      add_recipe(seeds_recipe) |>
      add_model(knn_spec) |>
      fit_resamples(resamples = seeds_vfold)

knn_tune <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
      set_engine("kknn") |>
      set_mode("classification")

knn_tune

knn_results <- workflow() |>
      add_recipe(seeds_recipe) |>
      add_model(knn_tune) |>
      tune_grid(resamples = seeds_vfold, grid = 10) |>
      collect_metrics()
knn_results

accuracies <- knn_results |> 
      filter(.metric == "accuracy")

accuracy_versus_k <- ggplot(accuracies, aes(x = neighbors, y = mean))+
      geom_point() +
      geom_line() +
      labs(x = "Neighbors", y = "Accuracy Estimate") +
      scale_x_continuous(breaks = seq(0, 14, by = 1)) +  # adjusting the x-axis
      scale_y_continuous(limits = c(0.4, 1.0)) # adjusting the y-axis
accuracy_versus_k
```
